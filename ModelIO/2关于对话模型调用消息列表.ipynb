1.正常调用大模型

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain_openai import ChatOpenAI
import os
import dotenv

#加载配置文件
dotenv.load_dotenv()

#获取模型的访问信息
os.environ['OPENAI_API_KEY'] = os.getenv('API_KEY')
os.environ['OPENAI_BASE_URL'] = os.getenv('BASE_URL')


chatModel = ChatOpenAI(
    base_url= os.environ['OPENAI_BASE_URL'],
    api_key=os.environ['OPENAI_API_KEY'],
    model_name='deepseek-chat',
)

response = chatModel.invoke('hello')
print(response.content)
```

2.常见的几种消息类型
SystemMessage,HumanMessage,AIMessage

```python
from langchain_openai import ChatOpenAI
import os
import dotenv
from langchain_core.messages import SystemMessage,HumanMessage,AIMessage

#加载配置文件
dotenv.load_dotenv()

#获取模型的访问信息
os.environ['OPENAI_API_KEY'] = os.getenv('API_KEY')
os.environ['OPENAI_BASE_URL'] = os.getenv('BASE_URL')

#定义消息类型
human_message = HumanMessage(content='java和python哪个好')
system_message = SystemMessage(content='你是一个java资深专家')
ai_message = AIMessage(content='java好')

chatModel = ChatOpenAI(
    base_url= os.environ['OPENAI_BASE_URL'],
    api_key=os.environ['OPENAI_API_KEY'],
    model_name='deepseek-chat',
)

messagess = [human_message, system_message, ai_message]

response = chatModel.invoke(messagess)

print(response.content)
```

3.大模型调用的几种方式
同步：invoke ，stream，batch
异步：ainvoke，astream，abatch

```python
from langchain_openai import ChatOpenAI
import os
import dotenv
from langchain_core.messages import SystemMessage,HumanMessage,AIMessage

#加载配置文件
dotenv.load_dotenv()

#获取模型的访问信息
os.environ['OPENAI_API_KEY'] = os.getenv('API_KEY')
os.environ['OPENAI_BASE_URL'] = os.getenv('BASE_URL')

#定义消息类型
human_message = HumanMessage(content='java和python哪个好')
system_message = SystemMessage(content='你是一个java资深专家')

chatModel = ChatOpenAI(
    base_url= os.environ['OPENAI_BASE_URL'],
    api_key=os.environ['OPENAI_API_KEY'],
    model_name='deepseek-chat',
    streaming=True
)

messagess = [human_message, system_message, ai_message]

for temp in chatModel.stream(messagess):
    print(temp.content,end='', flush=True)

print('流式输出结束了')
```
