# 如何创建PromptTemplate
1.通过实例方式创建  
2.通过from_template()的方式创建


# 通过实例的方式完成创建

```python
import os
import dotenv
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI


#1.加载配置文件
dotenv.load_dotenv()

#2.读取配置文件
os.environ['OPENAI_BASE_URL'] = os.getenv('BASE_URL')
os.environ['OPENAI_API_KEY'] = os.getenv('API_KEY')

#3.创建大模型对象
chatModel = ChatOpenAI(
     model="deepseek-chat"
)

#4.创建提示词模板
template = PromptTemplate(
    template = '你好,请问:{input}',
    input_variables = ['input']
)

# template = PromptTemplate('你好,请问:{input}')


#5.提示词赋值
templateValue = template.invoke({'input':'肥东圣泉中学'})

#6.调用大模型
chatModel.invoke(templateValue)
```

# 通过from_template()的方式创建

```python
import os
import dotenv
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI


#1.加载配置文件
dotenv.load_dotenv()

#2.读取配置文件
os.environ['OPENAI_BASE_URL'] = os.getenv('BASE_URL')
os.environ['OPENAI_API_KEY'] = os.getenv('API_KEY')

#3.创建大模型对象
chatModel = ChatOpenAI(
     model="deepseek-chat"
)

#4.创建模板
prompt = PromptTemplate.from_template('你好,请问:{input}')
# value = prompt.format_prompt({'input':'肥东马湖乡'})

value = prompt.format(input='肥东县马湖乡')


chatModel.invoke(value)
```
