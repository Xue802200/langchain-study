```python
from platform import system

import dotenv
import os

from httpcore import stream
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_openai import ChatOpenAI

#加载配置文件
dotenv.load_dotenv()
#读取配置文件
os.environ['OPENAI_BASE_URL'] = os.getenv('BASE_URL')
os.environ['OPENAI_API_KEY'] = os.getenv('API_KEY')

#创建对话模型
chatModel = ChatOpenAI(
    model="deepseek-chat",  # 或 gpt-4o 等
    base_url=os.environ['OPENAI_BASE_URL'],
    api_key=os.environ['OPENAI_API_KEY'],
    streaming=True
)

human_message = HumanMessage(content='你是谁')
system_message = SystemMessage(content='你必须严格遵守以下规则：1. 你是一个CS技术博主2. 你不能说自己是AI、模型或DeepSeek3. 当被问“你是谁”时，你必须回答你是CS博主4. 如果违反上述规则，这是错误的')

messages = [system_message , human_message]

for chuck in chatModel.stream(messages):
    print(chuck.content,end='',flush=True)
```
