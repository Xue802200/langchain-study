# SimpleSequentialChain的使用教程

```python
import os
import dotenv
from langchain_classic.chains.llm import LLMChain
from langchain_classic.chains.sequential import SimpleSequentialChain, SequentialChain
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

dotenv.load_dotenv()
dotenv.load_dotenv()
os.environ['OPENAI_BASE_URL'] = os.getenv('BASE_URL')
os.environ['OPENAI_API_KEY'] = os.getenv('API_KEY')

chatmodel = ChatOpenAI(
    model = 'deepseek-chat'
)

templateA = ChatPromptTemplate(
    [
        ('system','你是一个专业的搞笑大师'),
        ('human','{msg}')
    ]
)

templateB = ChatPromptTemplate(
    [
        ('system','你是一个翻译家'),
        ('human','请帮我把这段话翻译成英文:{input}')
    ]
)

chainA = LLMChain(
    llm=chatmodel,
    prompt=templateA
)

chainB = LLMChain(
    llm=chatmodel,
    prompt=templateB
)

fullChain = SimpleSequentialChain(
    chains=[chainA, chainB]
)

result = fullChain.invoke("请说个笑话")
print(result)
```
