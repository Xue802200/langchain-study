# LLMChain的使用

```python
import os
import dotenv
from langchain_classic.chains.llm import LLMChain
from langchain_classic.chains.sequential import SimpleSequentialChain
from langchain_core.prompts import ChatPromptTemplate, PromptTemplate
from langchain_openai import ChatOpenAI
from sqlalchemy.testing.provision import temp_table_keyword_args
from win32comext.adsi.demos.scp import verbose

dotenv.load_dotenv()
os.environ['OPENAI_BASE_URL'] = os.getenv('BASE_URL')
os.environ['OPENAI_API_KEY'] = os.getenv('API_KEY')

chatmodel = ChatOpenAI(
    model = 'deepseek-chat'
)
template = PromptTemplate.from_template(
    template='你好,请问:{input}'
)

chain = LLMChain(prompt=template,llm=chatmodel)
print(chain.invoke(input={'input':'你是谁'}))
```

# 顺序链SimpleSequentialChain的使用



```python
import os
import dotenv
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from langchain_classic.chains.sequential import SimpleSequentialChain

dotenv.load_dotenv()
os.environ['OPENAI_BASE_URL'] = os.getenv('BASE_URL')
os.environ['OPENAI_API_KEY'] = os.getenv('API_KEY')

chatmodel = ChatOpenAI(
    model = 'deepseek-chat'
)
templateA = PromptTemplate.from_template(
    template='你好,请问:{input}'
)
chainA = LLMChain(prompt=templateA,llm=chatmodel)

templateB = PromptTemplate.from_template(
    template='为什么这个笑话很好笑:{joke},要求最后输出结果在20字之内',
    verbose=True
)
chainB = LLMChain(prompt=templateB,llm=chatmodel)

fullChain = SimpleSequentialChain(
    chains = [chainA,chainB]
)

print(fullChain.invoke(input={'input':'请说个笑话'})['output'])
```
